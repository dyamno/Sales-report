```python

# Giả sử df là DataFrame của bạn đã được tạo trước đó
# Hiển thị hàng 4003 để xác nhận các giá trị
print("Hàng 4003:")
row_4003 = df.iloc[4003]
print(row_4003)

# Sử dụng điều kiện để lọc và loại bỏ các hàng có giá trị giống hàng 4003
condition = (df['Order ID'] == row_4003['Order ID']) & \
            (df['Product'] == row_4003['Product']) & \
            (df['Quantity Ordered'] == row_4003['Quantity Ordered']) & \
            (df['Price Each'] == row_4003['Price Each']) & \
            (df['Order Date'] == row_4003['Order Date']) & \
            (df['Purchase Address'] == row_4003['Purchase Address'])

# Lọc DataFrame để giữ lại các hàng không thỏa mãn điều kiện trên
df_cleaned = df[~condition].reset_index(drop=True)

# Hiển thị DataFrame sau khi loại bỏ các hàng có giá trị giống hàng 4003
print("\nDataFrame sau khi loại bỏ các hàng có giá trị giống hàng 4003:")
print(df_cleaned.head(10))  # Hiển thị 10 hàng đầu tiên để kiểm tra
```

```python

# Giả sử df là DataFrame của bạn đã được tạo trước đó

# Chuyển đổi cột 'Quantity Ordered' sang kiểu số, thay các giá trị không hợp lệ bằng NaN
df['Quantity Ordered'] = pd.to_numeric(df['Quantity Ordered'], errors='coerce', downcast='integer')

# Chuyển đổi cột 'Price Each' sang kiểu số, thay các giá trị không hợp lệ bằng NaN
df['Price Each'] = pd.to_numeric(df['Price Each'], errors='coerce', downcast='float')

# Loại bỏ các hàng chứa NaN sau khi chuyển đổi
df = df.dropna(subset=['Quantity Ordered', 'Price Each'])

# Hiển thị DataFrame sau khi làm sạch và chuyển đổi
print("DataFrame sau khi làm sạch và chuyển đổi:")
print(df.head(10))  # Hiển thị 10 hàng đầu tiên để kiểm tra
```

```python
df['Sales'] = df['Quantity Ordered'] * df['Price Each']
df.head()
```

```python
moving_column = df.pop('Sales')
df.insert(4,'Sales', moving_column)
df
```

```python
# Chuyển đổi cột 'Order Date' thành định dạng datetime
df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce')

# Tạo cột 'Month' từ cột 'Order Date'
df['Month'] = df['Order Date'].dt.month

# Tạo một DataFrame với tất cả các tháng
all_months = pd.DataFrame({'Month': range(1, 13)})

# Nhóm dữ liệu theo 'Month' và tính tổng 'Sales' cho mỗi tháng
sales_value = df.groupby('Month')['Sales'].sum().reset_index()

# Gộp dữ liệu của bạn với DataFrame tất cả các tháng
sales_value = all_months.merge(sales_value, on='Month', how='left').fillna(0)

# Kiểm tra số lượng phần tử trong sales_value
print("Số lượng phần tử trong sales_value:", len(sales_value))
print(sales_value)

# Vẽ biểu đồ cột
plt.bar(x=sales_value['Month'], height=sales_value['Sales'])
plt.xticks(range(1, 13))
plt.xlabel('Months')
plt.ylabel('Sales in USD')
plt.title('Monthly Sales')
plt.show()
```

```python
x = products_ls
y1 = all_products
y2 = prices

fig, ax1 = plt.subplots()
ax2  = ax1.twinx()
ax1.bar(x,y1, color= 'g')
ax2.plot(x, y2, 'b-')

ax1.set_xticklabels(products_ls, rotation=90, size=8)
ax1.set_xlabel('Products')
ax1.set_ylabel('Quantity Ordered', color='g')
ax2.set_ylabel('Price Each', color='b')
plt.show()
```
